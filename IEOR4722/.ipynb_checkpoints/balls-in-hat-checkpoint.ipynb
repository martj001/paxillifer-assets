```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def get_exp_return(n_black, n_white):
    # If there is no black ball, return = n_white
    if n_black == 0:
        exp_return[n_black, n_white] = n_white
        return None
    
    # If there is no white ball, you don't play
    if n_white == 0:
        exp_return[n_black, n_white] = 0
        return None
    
    if (n_black, n_white-1) not in exp_return:
        get_exp_return(n_black, n_white-1)
        
    if (n_black-1, n_white) not in exp_return:
        get_exp_return(n_black-1, n_white)
    
    exp_return[n_black, n_white] = calc_exp_return(n_black, n_white)

    return None

def calc_exp_return(n_black, n_white):
    """
    Requires exp_return[n_black - 1, n_white], exp_return[n_black, n_white - 1] exist
    """    
    n = n_black + n_white
    
    # draw white
    r_white = 1 + exp_return[(n_black, n_white-1)]
    p_white = n_white/n

    # draw black
    r_black = -1 + exp_return[(n_black-1, n_white)]
    p_black = n_black/n
    
    r = max(p_white*r_white + p_black*r_black, 0)
    
    return r
```

```python
# key: (n_black, n_white)
exp_return = {}

n_black = 10
n_white = 10
get_exp_return(n_black, n_white)

mat = np.zeros([n_black+1,n_white+1])

for key, val in exp_return.items():
    mat[key] = val
    
df_exp_return = pd.DataFrame(mat)
```
