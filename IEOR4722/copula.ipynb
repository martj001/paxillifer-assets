```python
from statsmodels.distributions.copula.api import CopulaDistribution, GaussianCopula, StudentTCopula, IndependenceCopula, GumbelCopula, FrankCopula, ClaytonCopula

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

# from statsmodels.graphics import utils
from matplotlib import cm
```

## Gumbel Copula

```python
ax_list = [
    [0,0],
    [0,1],
    [1,0],
    [1,1],
]

alpha_list = [
    1.001, 
    1.1,
    3,
    10,
]


fig, axs = plt.subplots(2,2, subplot_kw={"projection": "3d"},figsize=(15,10))

n_samples = 100
eps = 0.05
x, y = np.meshgrid(
    np.linspace(eps, 1 - eps, n_samples),
    np.linspace(eps, 1 - eps, n_samples)
)
points = np.vstack([x.ravel(), y.ravel()]).T

for i in range(4):
    alpha = alpha_list[i]
    copula = GumbelCopula(theta=alpha)
    
    z = copula.pdf(points).T.reshape(x.shape)
    z_min = min(z.flatten())
    z_max = max(z.flatten())
    
    ax = axs[ax_list[i][0], ax_list[i][1]]
    surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_zlim(0, z_max)
    
    # Add a color bar which maps values to colors.
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

    ax.azim = -100
    ax.set_title('alpha = ' + str(alpha))
    
plt.tight_layout()
```
